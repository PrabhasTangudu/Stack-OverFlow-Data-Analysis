```python
# Function to calculate the midpoint of an age range
def refined_midpoint(age_range):
    match = re.match(r'(\d+)-(\d+)', age_range)
    if match:
        low, high = map(int, match.groups())
        return (low + high) / 2
    return None

# Apply the function to the age data
survey_df['Age'] = survey_df['Age'].apply(refined_midpoint)
```

```python
import pandas as pd

def split_multicolumn(col_series):
    # Convert Series to DataFrame
    result_df = pd.DataFrame(index=col_series.index)
    
    # Iterate over non-null values
    for idx, value in col_series.dropna().items():
        # Split the value by ';' and iterate over each option
        for option in value.split(';'):
            # If the column does not exist, create it with False
            if option not in result_df.columns:
                result_df[option] = False
            # Set the cell to True
            result_df.at[idx, option] = True
    
    # Fill NaN values with False (for rows that had no options)
    result_df.fillna(False, inplace=True)
    
    return result_df

# Example usage:
# col_series = pd.Series(['a;b;c', 'a;c', 'b;c', None, 'a'])
# print(split_multicolumn(col_series))
```

#### Q: What are the most popular programming languages in 2023? 

To answer, this we can use the `LanguageWorkedWith` column. Similar to `DevType`, respondents were allowed to choose multiple options here.

It appears that a total of 51 languages were included among the options. Let's aggregate these to identify the percentage of respondents who selected each language.

```python
languages_worked_percentages = language_worked_df.mean().sort_values(ascending=False) * 100
languages_worked_percentages
```

```python
plt.figure(figsize=(12,12))
sns.barplot(x=languages_worked_percentages,y=languages_worked_percentages.index)
plt.title("Languages used in past year")
plt.xlabel("Count")
plt.ylabel("Languages");
```

What are the most common languages used by students? How does the list compare with the most common languages used by professional developers?

```python
print(survey_df.DevType)
```

```python
def get_most_common_languages(df, dev_type):
    languages = df[df['DevType'].str.contains(dev_type, na=False)]['LanguageHaveWorkedWith']
    languages_split = languages.str.split(';').explode()
    return languages_split.value_counts()
```

```python
# Most common languages used by students
student_languages = get_most_common_languages(survey_df, 'Student')
print(student_languages)
```

```python
plt.figure(figsize=(12,12))
plt.title("Most Commonly used ")
sns.barplot(x=student_languages,y=student_languages.index,hue=student_languages.index,palette='viridis')
plt.xlabel("Count")
plt.ylabel("Student Languages");
```

```python
pro_lang=get_most_common_languages(survey_df,"Developer Experience")
pro_lang
```

```python
# plotting for comparsion
plt.figure(figsize=(12,12))
sns.barplot(x=pro_lang,y=pro_lang.index,hue=pro_lang.index,palette='viridis')
plt.title("Languages used by professional")
plt.xlabel("Count")
plt.ylabel("Languages");
```

What are the most common languages among respondents who do not describe themselves as "Developer, front-end"?

```python
non_frontend_df=survey_df[~survey_df['DevType'].str.contains('Developer','front-end',na=False)]
non_frontend_languages=get_most_common_languages(non_frontend_df,"")
non_frontend_languages
```
